```python
import json
import pandas as pd
```

```python
# get a list of all cluster keywords from exhibition_clusters.json, institution_cluster_positions.json, and artist_cluster_positions.json, and write out a flat list of all possible keywords

with open("../../data/ui/exhibition_clusters.json", "r") as f:
    exhibition_clusters = json.load(f)

with open("../../data/ui/institution_cluster_positions.json", "r") as f:
    institution_cluster_positions = json.load(f)

with open("../../data/ui/artist_cluster_positions.json", "r") as f:
    artist_cluster_positions = json.load(f)

all_keywords = set()
for cluster in exhibition_clusters["exhibitions"]:
    all_keywords.update(cluster.get("labels", []))

for n in institution_cluster_positions:
    for cluster_data in n.get("cluster_positions", {}).values():
        all_keywords.update(cluster_data.get("keywords", []))

for n in artist_cluster_positions:
    for cluster_data in n.get("cluster_positions", {}).values():
        all_keywords.update(cluster_data.get("keywords", []))
```

```python
len(all_keywords)
```

```python
# sort and cast to a list
all_keywords = sorted(list(all_keywords))
keywords_df = pd.DataFrame(all_keywords, columns=["word"])
keywords_df.head()

keywords_df.to_csv("../../data/ui/cluster_keywords.csv", index=False)
```
