```python
import pandas as pd
```

### Load files

```python
today = pd.to_datetime('today').strftime('%m-%d')
artist_words = pd.read_csv(f"../../data/downloaded/processed/{today}/artist_words.csv")
artworks = pd.read_csv(f"../../data/downloaded/processed/{today}/artworks.csv")
institution_words = pd.read_csv(f"../../data/downloaded/processed/{today}/institution_words.csv")
```

```python
# construct new artist_texts that has everything in artist_words and all the artwork titles within artworks, keep the ids for source_idx lookup
artist_texts = (
    artist_words[["artist", "text", "id"]]
    .assign(is_artwork=False)
    .rename(columns={"id": "source_idx"})
    .to_dict("records")
)

# add each artwork title once with its artwork id for source_idx lookup
for _, artwork in artworks[["artist", "title", "id"]].dropna(subset=["title"]).iterrows():
    artist_texts.append({
        "artist": artwork["artist"],
        "text": artwork["title"],
        "source_idx": artwork["id"],
        "is_artwork": True,
    })

artist_texts_df = pd.DataFrame(artist_texts)
artist_texts_df["is_artwork"] = artist_texts_df["is_artwork"].fillna(False).astype(bool)
artist_texts_df.head()
```

```python
# do the same for institution words but add artwork descriptions and alt_text from artworks
institution_texts = (
    institution_words[["institution", "text", "id", "artist"]]
    .assign(is_artwork=False)
    .rename(columns={"id": "source_idx"})
    .to_dict("records")
)

for _, artwork in artworks[["institution", "description", "alt_text", "id", "artist"]].iterrows():
    if pd.notna(artwork["description"]):
        institution_texts.append({
            "institution": artwork["institution"],
            "text": artwork["description"],
            "source_idx": artwork["id"],
            "artist": artwork["artist"],
            "is_artwork": True,
        })
    if pd.notna(artwork["alt_text"]):
        institution_texts.append({
            "institution": artwork["institution"],
            "text": artwork["alt_text"],
            "source_idx": artwork["id"],
            "artist": artwork["artist"],
            "is_artwork": True,
        })

institution_texts_df = pd.DataFrame(institution_texts)
institution_texts_df["is_artwork"] = institution_texts_df["is_artwork"].fillna(False).astype(bool)
institution_texts_df.head()
```

```python
# export
artist_texts_df.to_csv(f"../../data/downloaded/processed/{today}/all_artist_texts.csv")
institution_texts_df.to_csv(f"../../data/downloaded/processed/{today}/all_institution_texts.csv")
```
